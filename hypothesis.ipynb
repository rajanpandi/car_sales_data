**Before encoding hypthoesis test**

```python
from scipy import stats
```

```python
def two_sample(d1,d2):
  m =[0,0]
  for i in range(31):
    sample1 = d1.sample(frac=0.03)
    sample2 = d2.sample(frac=0.04)
    t_test, p_value = stats.ttest_ind(sample1,sample2)
    if p_value < 0.04:
      #print("H0 Reject null hypothesis, Ha accepts alternate hypothesis ,  data is not normally distributed")
      m[1]+=1
    else:
      #print("H0 accept null hypothesis, Ha rejects alternate hypothesis,  data is normally distributed")
      m[0]+=1
    if m[0]>m[1]:
      return True
    elif m[0]<m[1]:
      return False
```

```python


def chi_square(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1]<0.075 else False
```

```python
import pandas as pd
from scipy import stats

def annova(d1, d2, df):
    # Get the unique categories in d2
    groups = df[d2].unique()

    # Skip if the categorical variable has less than 2 unique values
    if len(groups) < 2:
        return False  # Assuming no significant connection if there's only one group

    # Create a list of arrays, each containing the values of d1 for a specific category in d2
    data_groups = [df[d1][df[d2] == group] for group in groups]

    # Perform one-way ANOVA
    f_value, p_value = stats.f_oneway(*data_groups)

    # If p-value < 0.075, reject the null hypothesis
    return True if p_value >= 0.075 else False
```

```python
for i in df.columns:
    for j in df.columns:
        if i == j:
            continue  # Skip comparison with the same column
        elif i in coun and j in coun:
            # Clean and convert data before performing t-test
            if df[i].dtype == 'object':
                d1 = pd.to_numeric(df[i].str.replace(',', ''), errors='coerce')
            else:
                d1 = df[i]

            if df[j].dtype == 'object':
                d2 = pd.to_numeric(df[j].str.replace(',', ''), errors='coerce')
            else:
                d2 = df[j]

            result = two_sample(d1, d2)
            if result:
                print(i, j, "dependent")
            else:
                print(i, j, "independent")
        elif i in cata and j in cata:
            result = chi_square(df[i], df[j])
            if result:
                print(i, j, "dependent")
            else:
                print(i, j, "independent")
        elif i in coun and j in cata:
            # Clean and convert continuous data before performing ANOVA
            if df[i].dtype == 'object':
                df[i] = pd.to_numeric(df[i].str.replace(',', ''), errors='coerce')

            result = annova(i, j, df)
            if result:
                print(i, j, "dependent")
            else:
                print(i, j, "independent")
        elif i in cata and j in coun:
            # Clean and convert continuous data before performing ANOVA
            if df[j].dtype == 'object':
                df[j] = pd.to_numeric(df[j].str.replace(',', ''), errors='coerce')

            result = annova(j, i, df)
            if result:
                print(i, j, "dependent")
            else:
                print(i, j, "independent")
```
