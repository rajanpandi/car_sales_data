```python
import pandas as pd

def normalized_data(df, column_name):
    # Normalize the data from the specified column directly
    chennai_normalized_data = pd.json_normalize(df[column_name])

    return chennai_normalized_data

# Usage with your specific DataFrame and column
chennai_normalized_data = normalized_data(chennai_car_details1, 'new_car_detail')

# Display the normalized DataFrame
chennai_normalized_data
```

```python
import ast
import pandas as pd

def normalized1_data(df, column_name):
    # Ensure the specified column is a dictionary (convert from string if needed)
    df[column_name] = df[column_name].apply(ast.literal_eval)

    # Extract the 'top' data and create columns from the keys
    top_data = df[column_name].apply(lambda x: {item['key']: item['value'] for item in x['top']})

    # Convert the dictionary of 'top' data to a DataFrame
    return pd.DataFrame(top_data.tolist())

# Usage example with chennai_car_overview1 DataFrame
chennai_car_overview1_normalized_data = normalized1_data(chennai_car_overview1, 'new_car_overview')

# Display the resulting DataFrame
chennai_car_overview1_normalized_data
```

```python
data1.columns
```

```python
chennai_car_feature=data1['new_car_feature']
chennai_car_feature1=pd.DataFrame(chennai_car_feature)
```

```python
import ast
import pandas as pd

def normalized2_data(df, column_name):

    # Ensure the specified column is a dictionary (convert from string if needed)
    if df[column_name].apply(type).eq(str).any():
        df[column_name] = df[column_name].apply(ast.literal_eval)

    # Initialize an empty dictionary to hold column data
    columns = {}

    # Process the 'top' data
    df['top'] = df[column_name].apply(lambda x: x.get('top', []))
    columns['Features'] = df['top'].apply(len)

    # Process the 'data' field
    df['data'] = df[column_name].apply(lambda x: x.get('data', []))
    for entry in df['data'].dropna().values[0]:
        heading = entry['heading']
        list_length = len(entry['list'])
        columns[f'{heading}'] = [list_length] * len(df)

    # Convert the dictionary to a DataFrame
    chennai_car_feature_normalized_data = pd.DataFrame(columns)
    return chennai_car_feature_normalized_data

# Usage example with chennai_car_feature1 DataFrame
chennai_car_feature_normalized_data = normalized2_data(chennai_car_feature1, 'new_car_feature')

# Display the resulting DataFrame
chennai_car_feature_normalized_data
```

```python
chennai_car_specification=data1['new_car_specs']
chennai_car_specification1=pd.DataFrame(chennai_car_specification)
chennai_car_specification1
```

```python
import ast
import pandas as pd

def normalized4_data(df, column_name):
    # Ensure the specified column is a dictionary (convert from string if needed)
    df[column_name] = df[column_name].apply(ast.literal_eval)

    # Normalize the data from the specified column
    normalized_data = df[column_name].apply(
        lambda specs: {
            **{item['key']: item['value'] for item in specs.get('top', [])},
            **{item['key']: item['value'] for section in specs.get('data', []) for item in section.get('list', [])}
        }
    )

    # Convert to DataFrame
    chennai_car_specification1_normalized_data = pd.json_normalize(normalized_data)

    return chennai_car_specification1_normalized_data

# Use the function and assign the result to a variable
chennai_car_specification1_normalized_data = normalized4_data(chennai_car_specification1, 'new_car_specs')

# Display the resulting DataFrame
chennai_car_specification1_normalized_data
```

```python
chennai_full_data=pd.concat([chennai_normalized_data,chennai_car_overview1_normalized_data,chennai_car_feature_normalized_data,chennai_car_specification1_normalized_data],axis=1)
```
