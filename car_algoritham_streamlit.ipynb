```python
x=df.drop(["price"],axis=1)
y=df["price"]
```

```python
#standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #(data - mean(data))/ Sd(data)
scaler.fit(x)
x_scaler = scaler.transform(x)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaler,y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression().fit(x_train,y_train)
y_pred=model1.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor().fit(x_train,y_train)
y_pred=model2.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.svm import SVR
model3 = SVR().fit(x_train,y_train)
y_pred=model3.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsRegressor
model4 = KNeighborsRegressor().fit(x_train,y_train)
y_pred=model4.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))#priority 2
print(r2_score(y_test,y_pred)) #priority 1
```

```python
from xgboost import XGBRegressor
model5 = XGBRegressor().fit(x_train,y_train)
y_pred=model5.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))#priority 2
print(r2_score(y_test,y_pred)) #priority 1
```

```python
from sklearn.tree import DecisionTreeRegressor
# Initialize the regressor
regressor = DecisionTreeRegressor()

# Fit the model on your training data
regressor.fit(x_train, y_train)

# Make predictions on the test data
y_pred = regressor.predict(x_test)

print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))#priority 2
print(r2_score(y_test,y_pred)) #priority 1
```
